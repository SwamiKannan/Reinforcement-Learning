```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('episode_loss.txt.npy','rb') as f:
    DQN_overfit_loss=np.load(f,allow_pickle=True)
```

```python
with open('episode_loss_large.txt.npy','rb') as f:
    DQN_fit_loss=np.load(f,allow_pickle=True)
```

```python
with open('epside_rewards.txt.npy','rb') as f:
    DQN_overfit_reward=np.load(f,allow_pickle=True)
```

```python
with open('epside_rewards_large.txt.npy','rb') as f:
    DQN_fit_reward=np.load(f,allow_pickle=True)
```

```python
with open('DDQN_episode_loss.txt.npy','rb') as f:
    DDQN_overfit_loss=np.load(f,allow_pickle=True)
```

```python
with open('DDQN_epside_rewards.txt.npy','rb') as f:
    DDQN_overfit_reward=np.load(f,allow_pickle=True)
```

```python
with open('DDQN_large_episode_loss_large.npy','rb') as f:
    DDQN_fit_loss=np.load(f,allow_pickle=True)
```

```python
with open('DDQN_large_epside_rewards_large.npy','rb') as f:
    DDQN_fit_reward=np.load(f,allow_pickle=True)
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(2000),DQN_overfit_reward,color='#ff9c00',label='DQN_overfit_reward')
plt.plot(range(2000),DQN_fit_reward,color='#2400ff',label='DQN_fit_reward')
plt.plot(range(2000),DDQN_overfit_reward,color='#00ff12',label='DDQN_overfit_reward')
plt.plot(range(2000),DDQN_fit_reward,color='#ff00de',label='DDQN_fit_reward')
plt.axhline(y=200,linewidth=1,color='red')
plt.xlabel('Episode number')
plt.ylabel('Rewards for episode')
plt.title('Rewards vs. Episode number')
plt.legend()
plt.show()
```

It isn't very clear but mainly we see that initially the DQN model takes a little longer to peak but towards around 1750 episodes, DDQN overfit model get far more volatility 

```python
plt.figure(figsize=(15,5))
plt.plot(range(2000),DQN_fit_reward,color='#2400ff',label='DQN_fit_reward')
plt.plot(range(2000),DQN_overfit_reward,color='#ff9c00',label='DQN_overfit_reward')
plt.axhline(y=200,linewidth=1,color='red')
plt.xlabel('Episode number')
plt.ylabel('Rewards for episode')
plt.title('Rewards vs. Episode number')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(2000),DDQN_overfit_reward,color='#00ff12',label='DDQN_overfit_reward')
plt.plot(range(2000),DDQN_fit_reward,color='#ff00de',label='DDQN_fit_reward')
plt.axhline(y=200,linewidth=1,color='red')
plt.xlabel('Episode number')
plt.ylabel('Rewards for episode')
plt.title('Rewards vs. Episode number')
plt.legend()
plt.show()
```

Increasing the size of the model, the batch size and the buffer size reduces the voliatility of the training in more intervals

```python
plt.figure(figsize=(15,5))
plt.plot(range(2000),DQN_overfit_loss,color='#ff9c00',label='DQN_overfit_loss')
plt.plot(range(2000),DQN_fit_loss,color='#2400ff',label='DQN_fit_loss')
plt.plot(range(2000),DDQN_overfit_loss,color='#00ff12',label='DDQN_overfit_loss')
plt.plot(range(2000),DDQN_fit_loss,color='#ff00de',label='DDQN_fit_loss')
plt.xlabel('Episode number')
plt.ylabel('Loss for episode')
plt.title('Losses vs. Episode number')
plt.legend()
plt.show()
```

the DDQN loss shows the least volatility and the least amount of loss across all cycles. The overfit model of DDQN did show quite a few spikes including a loss of near 7000

```python
DDQN_overfit_regularize_loss=DDQN_overfit_loss.copy()
DDQN_overfit_regularize_loss[np.where(DDQN_overfit_loss==max(DDQN_overfit_loss))]=np.mean(DDQN_overfit_loss)
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(2000),DQN_overfit_loss,color='#ff0000',label='DQN_overfit_loss')
plt.plot(range(2000),DQN_fit_loss,color='#2400ff',label='DQN_fit_loss')
plt.plot(range(2000),DDQN_overfit_regularize_loss,color='#00ff12',label='DDQN_overfit_regularize_loss')
plt.plot(range(2000),DDQN_fit_loss,color='#ff00de',label='DDQN_fit_loss')
plt.xlabel('Episode number')
plt.ylabel('Loss for episode')
plt.title('Losses vs. Episode number')
plt.legend()
plt.show()
```

We can see that DQN typically tends to have the right direction of improvement but far more volatility. DQN fit and overfit seem to however have a near lockstep improvement in performance
